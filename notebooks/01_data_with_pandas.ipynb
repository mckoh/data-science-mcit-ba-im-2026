# Arbeiten mit Daten in `pandas`

```python
import pandas as pd
```

```python
df = pd.read_csv(
    "../data/eistage.csv",
    sep=";",
    skiprows=2,
    nrows=70,
    usecols=[3,5,6],
    parse_dates=["year"],
    na_values=[1900, -999, -999],
    names=["year", "heat", "ice"],
    index_col=0
)
```
